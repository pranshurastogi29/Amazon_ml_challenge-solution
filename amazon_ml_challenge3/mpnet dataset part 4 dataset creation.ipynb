```python
import numpy as np
import pandas as pd
import torch
import gc
from sentence_transformers import SentenceTransformer
```

```python
test = pd.read_csv('../input/amazon-clean-dataset/test.csv')
model = SentenceTransformer('stsb-mpnet-base-v2')
embeddings = model.encode(test['All'], show_progress_bar=True)
del model , test
```

```python
np.save('test_embeddings.npy',embeddings)
```

```python
del embeddings
```

```python
df = pd.read_csv('../input/amazon-clean-dataset/data.csv')
model = SentenceTransformer('stsb-mpnet-base-v2')
df_split = df['All'].values[1500000:2000000]
embeddings = model.encode(df_split, show_progress_bar=True)
del model , df
```

```python
np.save('data_embeddings4.npy',embeddings)
```
